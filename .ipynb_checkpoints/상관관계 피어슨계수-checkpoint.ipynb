```python
# pandas로 불러옴
df=pd.read_csv('./diabetes.csv')
df  
# df.iloc
x = df.iloc[:,0:8]
y = df.iloc[:,8]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
import numpy
import tensorflow as tf
df_data = df[["Pregnancies","Glucose","BMI","Age"]]
df_label = df["Outcome"]

# 모델을 설정합니다.
model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
X_train, X_test, y_train, y_test = train_test_split(df_data, df_label, test_size=0.1, random_state=42)

# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics=['accuracy'])
history=model.fit(X_train,y_train,epochs=200, batch_size=10)
```

```python
print('\n Accuracy: %.4f' %(model.evaluate(df_data,df_label)[1]))
```

```python
import matplotlib.pyplot as plt

# plt.plot(history.history)함수:epoch에 따른 정확도와 손실도를 기록하는 속성 
plt.plot(history.history['loss'], label="loss")
plt.plot(history.history['accuracy'],label="loss")
plt.legend()

plt.show()
```

```python
# X_test의 길이가 77개
len(X_test)
```

```python
# X_test에 대해 predict예측을 함. 첫번째는 당뇨가 있다, 두번째는 당뇨가 없다 등등으로 예측함 
results = model.predict(X_test)
result_array=[]

for i in range(len(results)):
    if results[i][0] < 0.5:
        result_array.append(0)
    else:
        result_array.append(1)
X_test['Outcome']=y_test
X_test['Predict']=result_array

X_test
```

```python
from sklearn.metrics import accuracy_score,recall_score, precision_score, f1_score
```

```python
acc = accuracy_score(X_test['Outcome'],X_test['Predict'])
recall = recall_score
precision_score = precision_score(X_test['Outcome'], X_test['Predict'])

f1 = f1_score(X_test['Outcome'], X_test['Predict'])
```

cross-entropy는 확률에 log함수 씌운다음 1,0을 곱함
여기서는 binary이기 때문에 1이라는 값에 집중. 0.5이상이면 1로 보고 0.5 미만이면 0으로 봄
log함수취한 값 끼리의 오차를 계산. 
만약에 0이라는 숫자에 집중하면 숫자를 바꿔줘야함. 

붓꽃 데이터에서 잎,줄기,뿌리에 각각 0,1,2로 나오면(다중분류의 classification인 경우)
-> One-Hot encoding을 통해 0,1로 바꿔줘야 함


backpropagation을 통해 weight와 bia를 조절해 주고 그래서 다시 propagatio을 통해 update시킨 값을 출력
오차를 줄여주기 위해 계속 반복. 그래서 

x축을 보면 100이상이 되면 더이상 변동이 없으니까 100에서 멈추려고 함 -> early stopping

다중 분류의 문제에서는 어디를 줄일지를 고민해서 cross-entropy를 줄여야 함.
loss값, roc값 등등 여러가지를 이해해야 함
